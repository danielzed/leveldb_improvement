```python
import re
```

```python
pattern = re.compile(".*key:(\d+)")
data_map = {}
```

```python
with open("data","r") as f:
    line = f.readline()
    while(line):
        result = pattern.match(line)
        if(result):
            keystr = result.group(1)
            if(keystr not in data_map):
                data_map[keystr] = 0
            data_map[keystr] = data_map[keystr]+1
        line = f.readline()
print(data_map)
```
